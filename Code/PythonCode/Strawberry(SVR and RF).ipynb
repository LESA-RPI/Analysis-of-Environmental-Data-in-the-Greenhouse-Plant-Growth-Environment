```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data1 = pd.read_excel('StrawberryOrganizedScaled.xlsx')
data1
```

```python
target = pd.read_excel('Strawberry_Target_CA.xlsx')
y_train= target.iloc [:22, 0]
y_train
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
x_train= data1.iloc [:22, :]
y_train= target.iloc [:22, 5]
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train, y_train.values.ravel())
x_test = data1.iloc [22 :, :]
y_test = target.iloc [22 :, 5]
prediction = regressor.predict(x_test)
mse = mean_squared_error(y_test, prediction)
rmse = mse**.5
#print(mse)
print(rmse)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
x_train= data1.iloc [:22, :]
y_train= target.iloc [:22, 5]
regressor = RandomForestRegressor(n_estimators = 100)
regressor.fit(x_train, y_train.values.ravel())
x_test = data1.iloc [22 :, :]
y_test = target.iloc [22 :, 5]
prediction = regressor.predict(x_test)
mse = mean_squared_error(y_test, prediction)
rmse = mse**.5
#print(mse)
print(rmse)
```
